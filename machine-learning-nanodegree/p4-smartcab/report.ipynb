>**Action Selection Method**

>There are multiple action selection policies worth mentioning. Boltzmann, Îµ-greedy and greedy selection. In many simple cases a Îµ-greedy machanism can be chosen where the Îµ value can be kept fixed for all number of iterations. Usually, the amount exploration done by the agent should diminish over time, meaning that the policy should get increasingly greedy. This can be achieved making Îµ approach 0 as the number of trials t increase. A simple form of this could by Îµ = 1/t. The Boltzmann policy is "softmax" method that uses the Gibbs distribution to estimate the probability of taking an action "a" in a given state "s". In the greedy policy, the agent selects the most interesting actions (as an evaluation function) in a given state. In this case, the agent does not explore all actions.

>I think an increasingly greedy policy is the best fit in this case. As the agent needs to learn a fairly large amount of state, action pair values it makes sense to let the agent explore a lot in the beginning before making it more greedy once it has a more complete view of the world. Having a low exploration rate towards the end will also stop the agent from taking any negative reward actions. I have implemented an *epsilon-decay* action selection function that selects a random action with probability $e^{-Î³t}$ and otherwise the optimal action according to the policy.

>** Îµ-decay**

>**Parameter Tuning**

> Crucial to successful reinforcement learning is the careful setting of the three meta-parameters Î±; Îµ and Î³.

> For this particular problem we know that the a given (state, action) pair always will give the same results. We can therefore deduce that the optimal learning rate is 1, as this is a fully deterministic environment. <a href="https://en.wikipedia.org/wiki/Q-learning#Learning_rate">[1]</a>

> The dicount factor Î³ determines the importance of the future estimated reward. A factor of 0 will make the agent "short-sighted" by only considering current rewards, while a factor close to 1 will make it strive for a long-term high reward. As we know that the optimal action doesn't depend on the future. The optimal discount factor value is therefore 0.

> This leaves us with Îµ, for which we need to find the optimal value. There are multiple ways to go about tuning the parameters. However, due to the long running time of the simulation (which can be decreased by not displaying graphics) search algorithms such as evolutionary algorithms will take a long time to run, it would also require me to write some intricate fitness function. I have therefore chosen the easy route, namely to select the optimal parameters through trial and error.
